**Preprocessing For The Model**

Defining The Input And The Target Data

```python
input_col = list(df.columns)[1:-1]
target_col = 'SalePrice'
```

```python
x = df[input_col].copy()
y = df[target_col].copy()
```

```python
y
```

Splitting the Numerical Data and Categorical Data

```python
numeric_cols = x.select_dtypes(include=np.number).columns.tolist()
categorical_cols = x.select_dtypes('object').columns.tolist()
```

```python
numeric_cols
```

```python
categorical_cols
```

```python
x[categorical_cols].nunique()
```

Filling The Missing Values In The Numerical Data

```python
x[numeric_cols].isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute=SimpleImputer(strategy='mean')
```

```python
impute.fit(df[numeric_cols])
```

```python
x[numeric_cols] = impute.transform(x[numeric_cols])
```

```python
x[numeric_cols].isnull().sum()
```

Performing One Hot Encoder To The Numerical Data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Encoder=OneHotEncoder( handle_unknown='ignore')
```

```python
Encoder.fit(x[categorical_cols])
```

```python
# Fit and transform
encoded_data = Encoder.transform(x[categorical_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=Encoder.get_feature_names_out(categorical_cols))

# Drop original categorical columns
x = x.drop(columns=categorical_cols)

# Concatenate the new one-hot encoded columns
x = pd.concat([x, encoded_df], axis=1)
```

```python
x
```

Scaling The Numerical Values By Standard Scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler= StandardScaler()
```

```python
Scaler.fit(x[numeric_cols])
x[numeric_cols]=Scaler.transform(x[numeric_cols])
```

```python
x
```

Spliting The Data into Training And Testing Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Finally Training The Data

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge()
```

```python
model.fit(X_train,Y_train)
```

Evaluating The Model

```python
prediction = model.predict(X_test)
prediction1 = model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(Y_test, prediction)
rsme=np.sqrt(mse)


print(f"Test Data Root Mean Squared Error: {rsme}")
```

```python
from sklearn.metrics import mean_squared_error
mse1 = mean_squared_error(Y_train, prediction1)
rsme1=np.sqrt(mse1)


print(f"Train Data Root Mean Squared Error: {rsme1}")
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(Y_test, prediction) * 100
mape2 = mean_absolute_percentage_error(Y_train, prediction1) * 100
print(f"Error Percentage On Test Data: {mape:.2f}%")
print(f"Error Percentage On Train Data: {mape2:.2f}%")
```
