**Importing Dependencies**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
```

**Loading up the datasets and doing pre processing on it.**

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.sample(10)
```

As the Column "id" is not relevant to for the training i am gonna drop that column

```python
df.drop('Id', axis=1, inplace=True)
```

```python
input_col = list(df.columns)[1:-1]
target_col = 'SalePrice'
```

```python
x=df[input_col]
y=df[target_col]
```

Splitting the data into categorical and numerical


```python
numeric_cols = x.select_dtypes(include=np.number).columns.tolist()
categorical_cols = x.select_dtypes('object').columns.tolist()
```

Handling the missing values

```python
x[numeric_cols].isna().sum().sort_values(ascending=False).head(10)
```

```python
from sklearn.impute import SimpleImputer
```

```python
Impute = SimpleImputer(strategy='mean')
```

```python
Impute.fit(x[numeric_cols])
```

```python
Impute.transform(x[numeric_cols])
```

```python
x[numeric_cols].isna().sum().sort_values(ascending=False).head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
Encoder=OneHotEncoder( handle_unknown='ignore')
Encoder.fit(x[categorical_cols])
```

```python
# Fit and transform
encoded_data = Encoder.transform(x[categorical_cols])

# Convert to DataFrame
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=Encoder.get_feature_names_out(categorical_cols))

# Drop original categorical columns
x = x.drop(columns=categorical_cols)

# Concatenate the new one-hot encoded columns
x = pd.concat([x, encoded_df], axis=1)
```

Scaling the numerical data

```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
Scaler.fit(x[numeric_cols])
```

```python
x[numeric_cols]=Scaler.transform(x[numeric_cols])
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Lets Create The Model (Decision Tree)

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor()
model.fit(X_train, Y_train)
```

lets evaluate the model


```python
pred=model.predict(X_train)
pred2=model.predict(X_test)
```

```python
print(Y_train.shape)
print(X_train.shape)
```

```python
from sklearn.metrics import mean_squared_error
mse1 = mean_squared_error(Y_train, pred)
mse2 = mean_squared_error(Y_test, pred2)
rsme1=np.sqrt(mse1)
rsme2=np.sqrt(mse2)
print(f"Train Data Root Mean Squared Error: {rsme1}")
print(f"Test Data Root Mean Squared Error: {rsme2}")
```

As you can see that the loss on train data is 0
while the loss on test data is arounf 4k,
this means the model has overfitted.
Lets try to fix this overfitting by hyperparameter tunning

```python
def loss_calculate (pred,pred2):
  mse1 = mean_squared_error(Y_train, pred)
  mse2 = mean_squared_error(Y_test, pred2)
  rsme1=np.sqrt(mse1)
  rsme2=np.sqrt(mse2)
  print(f"Train Data Root Mean Squared Error: {rsme1}")
  print(f"Test Data Root Mean Squared Error: {rsme2}")
```

lets create a function to test out different parameters and their different values to minimize the loss.

```python
def test_params(**params):
    model = DecisionTreeRegressor(random_state=42 , **params).fit(X_train, Y_train)
    pred = model.predict(X_train)
    pred2 = model.predict(X_test)
    loss_calculate(pred,pred2)
```

```python
test_params(max_depth=10)
```

```python
test_params(max_depth=15)
```

```python
test_params(max_depth=5)
```

```python
test_params(max_depth=6)
```

So we get the minimum loss with max_depth = 6

```python
test_params(max_depth=6,max_leaf_nodes=2**20)
```

So this is the minimum loss that we got to with hyperparameter tunning.
**Now lets try Random Forest Instead of Decision Tree**

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
model.fit(X_train, Y_train)
```

```python
model = RandomForestRegressor(random_state=42 ).fit(X_train, Y_train)
pred = model.predict(X_train)
pred2 = model.predict(X_test)
loss_calculate(pred,pred2)
```

thats a really good start, now lets tune hyperparameters and lets see how much we can minimize the loss

```python
def test_params2(**params):
    model = RandomForestRegressor(random_state=42 , **params).fit(X_train, Y_train)
    pred = model.predict(X_train)
    pred2 = model.predict(X_test)
    loss_calculate(pred,pred2)
```

```python
test_params2(n_estimators=250,max_depth=20,max_leaf_nodes=2**10)
```

**So we got to 28761 Root mean Squared Error with some hyperparameter tunning.**
